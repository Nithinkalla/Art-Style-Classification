```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import numpy as np
import matplotlib.pyplot as plt

# Load the saved model
model = load_model("best_model.keras")

# Path to a single image
image_path = r"C:\Users\NITHIN\Desktop\334\test\Baroque\boarque.jpg"

# Define class labels (ensure these match the order used in training)
class_labels = ["Baroque", "Cubism", "Expressionism", "Realism", "Romanticism"]  # Replace with actual labels

# Function to preprocess, predict, and display a single image
def predict_and_show_image(image_path):
    # Load and preprocess the image
    img = load_img(image_path, target_size=(128, 128))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions to match the input shape
    img_array /= 255.0  # Normalize as done during training

    # Make a prediction
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)[0]
    confidence = np.max(predictions)
    predicted_label = class_labels[predicted_class]

    # Display the image with the prediction
    plt.imshow(load_img(image_path))  # Load image in original colors for display
    plt.axis('off')
    plt.title(f"Predicted style: {predicted_label}")
    plt.show()

# Predict and display the single image
predict_and_show_image(image_path)
```
